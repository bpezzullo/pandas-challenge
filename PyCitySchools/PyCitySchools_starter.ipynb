```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

```python
school_count = school_data_complete["school_name"].nunique()
student_count = school_data_complete["student_name"].nunique()
math_pass = (school_data_complete[school_data_complete["math_score"] > 70].count() /student_count) * 100
reading_pass = (school_data_complete[school_data_complete["reading_score"] > 70].count() /student_count) * 100
total_pass = (school_data_complete[(school_data_complete["math_score"] > 70) & (school_data_complete["reading_score"] > 70)].count() /student_count) * 100


district_summary = pd.DataFrame({"School Count":school_data_complete["school_name"].nunique(),
                                "Student Count" : school_data_complete["student_name"].nunique(),
                                "Total Budget" : school_data_complete["budget"].sum(),
                                "Average Math Score" : school_data_complete["math_score"].mean(),
                                "Average Reading Score" : school_data_complete["reading_score"].mean(),
                                "Percentage of passing Math scores" : math_pass.max(),
                                "Percentage of passing Reading scores" : reading_pass.max(),
                                "Percentage overall passing" : total_pass.max()                                
                                },index=[0])
district_summary["Percentage of passing Reading scores"] = district_summary["Percentage of passing Reading scores"].map("{:.2f} %".format)
district_summary["Percentage of passing Math scores"] = district_summary["Percentage of passing Math scores"].map("{:.2f} %".format)
district_summary["Percentage overall passing"] = district_summary["Percentage overall passing"].map("{:.2f} %".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:.2f}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,}".format)
district_summary
```
